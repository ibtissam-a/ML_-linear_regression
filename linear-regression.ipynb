# **linear regression with single variable**

```python
import numpy as np  
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# create a dictionnary
gain = {"nb_costumer" : [50,80,100,140,190,230,300,350,400,550,600,750],
         "sels" : [2000,2500,3000,3300,3750,4000,4900,5500,6500,8000,9950,10000]}
```

```python
# create our gain dataframe
df_gain = pd.DataFrame(gain)
df_gain
```

```python
# plot our data
plt.xlabel('nb_costumer')
plt.ylabel('sels')
plt.scatter(df_gain.nb_costumer, df_gain.sels, marker="*", color= "blue")
```

```python
# create our linear regression model 
linear_reg = linear_model.LinearRegression()
# fit our linear regression model 
linear_reg.fit(df_gain[['nb_costumer']], df_gain.sels)
```

```python
linear_reg.predict([[300]])
```

```python
# sels = m * nb_costumer + b  
# m : is the slop or the gradient 
# b : is the intercept
m = linear_reg.coef_
print("the slop value is : ", m)
b = linear_reg.intercept_
print("the intercept value is : ", b)
```

```python
# plot the predictetd values with error reduction
plt.xlabel('nb_costumer')
plt.ylabel('sels')
plt.plot(df_gain.nb_costumer, linear_reg.predict(df_gain[['nb_costumer']]), color = 'green')
```

```python
df_gain['pred'] = linear_reg.predict(df_gain[['nb_costumer']])
df_gain
```

```python
# we will create our test dictionnary
test = {"nb_costumer": [5, 10, 15, 25, 52, 106, 208, 700, 900]}
df = pd.DataFrame(test)
df.head(2)
```

```python
pred = linear_reg.predict(df)
df["sels"] = pred
```

```python
df.head(5)
```

```python
df.to_csv("gain.csv", index = False)
```

# **linear regression with multiple variables **

```python
df = pd.read_csv('../input/hiring/hiring.csv')
df
```

```python
df.rename(columns = {'test_score(out of 10)': 'test_sc',
                     'interview_score(out of 10)' : 'interview_sc', 
                     'salary($)' : 'salary'}, inplace = True)
df
```

```python
# replace the string values of the experience column with integers 
df['experience'] = ['NaN','NaN', 5,2,7,3,10,11]
df
```

```python
df.iloc[6,1] = df.test_sc.median()
df.iloc[0,0] = df.experience.median().round()
df.iloc[1,0] = df.experience.median().round()
df
```

```python
reg_model = linear_model.LinearRegression()
reg_model.fit(df[['experience', 'test_sc', 'interview_sc']], df.salary)
```

```python
reg_model.coef_
```

```python
reg_model.intercept_
```

```python
print("the salary for a condidate with 2 years experince and 9 test_sc and 6 interview_sc is :",  reg_model.predict([[2,9,6]]))
print("the salary for a condidate with 2 years experince and 9 test_sc and 6 interview_sc is :",  reg_model.predict([[12,10,10]]))
```
